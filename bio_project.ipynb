## 1. По файлам .feature_table посчитаем длину, количество аннотированных генов, процент покрытия

```python
def get_stats(filename, what_count):
    feature_table = pd.read_csv(filename, sep='\t')
    leng = 0
    for item in SeqIO.parse(f"{filename[0:-17]}genomic.fna", "fasta"):
        leng += len(item.seq)
    annotated = feature_table[(feature_table['# feature']=='gene') & (feature_table['seq_type']=='chromosome')]
    annotated = annotated[annotated['end'] > annotated['start']]
    annotated_leng = sum(annotated['end'] - annotated['start'])
    if what_count == 1:
      print(f"Количество аннотированных генов для {filename[0:-17]}:", annotated.shape[0])
    elif what_count == 2:
      print(f"Длина аннотированной последовательности {filename[0:-18]}: ", annotated_leng)
    elif what_count == 3:
      print(f"Длина  последовательности {filename[0:-18]}: ", leng)
    else:
      print(f"Доля аннотированных генов {filename[0:-18]}: ", annotated_leng / leng * 100)
```

```python
import os
import subprocess
import tempfile
from pathlib import Path
from subprocess import DEVNULL, PIPE

import pandas as pd

ZH_EXECUTABLE = Path("/content/zhunt3")
assert ZH_EXECUTABLE.is_file()


def zhunt(query: str, windowsize: int = 6, minsize: int = 3, maxsize: int = 6):
    print(set(query))
    for symb in ['M', 'R', 'K', 'W', 'Y', 'S', 'V']:
      query = query.replace(symb, 'N')
    assert set(query).issubset({"A", "C", "G", "T", "N"})
    fd, temp = tempfile.mkstemp()
    os.close(fd)
    with open(temp, 'w') as stream:
        stream.write(query)

    subprocess.run(
        [ZH_EXECUTABLE, 
         str(windowsize), str(minsize), str(maxsize), temp],
        check=True, stdout=PIPE, stderr=DEVNULL,
        input=query, encoding='ascii'
    )
    with open(temp + ".Z-SCORE", 'r') as stream:
        df = pd.read_csv(stream,
                         names=['Start', 'End', 'nu-1', 'nu-2', 'nu-3', 
                                'ZH-Score', 'Sequence', 'Conformation'],
                         skiprows=1, sep='\s+')
    os.remove(temp)
    os.remove(temp + ".Z-SCORE")
    return df[['Start', 'End', 'ZH-Score', 'Sequence', 'Conformation']]
```

```python
from Bio import SeqIO
list_df = []
file_names = ['GCF_004798665.1_ASM479866v1_feature_table.txt', 'GCF_002076915.1_ASM207691v1_feature_table.txt', 
              'GCF_013177295.1_ASM1317729v1_feature_table.txt', 
               'GCF_016127855.1_ASM1612785v1_feature_table.txt', 'GCF_016127955.1_ASM1612795v1_feature_table.txt']
for filename in file_names:
    with open(filename[0:-17] + "genomic.fna", 'r') as f:
        seq = list(SeqIO.parse(f, format='fasta'))
    for s in seq:
        if (s.description).__contains__('chromosome'):
            print(f"{s.description} \n\t=> {len(s)}")
            dna = zhunt(str(s.seq))
            list_df.append(dna.copy())
```

```python
def make_bed(data, fileName, genome_file):
  
  to_write = data.copy()
  to_write.insert(0, 'chr', str(genome_file).split('.')[0])
  to_write['name'] =to_write.chr.astype(str) +'_'+ to_write.Start.astype(str)
  to_write[['chr','Start','End','name','ZH-Score']].to_csv(fileName, sep = '\t', header=False, index=False)
```

# Ассоциируем предсказанные участки Z-DNA с промотерами генов

```python
import numpy as np
```

```python
ind_name = 0
cords_ = []
for name in names:
    file_ = pd.read_csv(f"{name}_feature_table.txt", sep='\t')
    f = pd.DataFrame(columns=['name', 'start', 'end', 'n', "protein", "strand"])
    idx = 0
    check_gene = False
    for i in range(len(file_)):
        if file_.loc[i, "# feature"] == "gene":
            if file_.loc[i, "strand"] == '+':
                position = file_.loc[i, "start"]
            elif file_.loc[i, "strand"] == '-':
                position = file_.loc[i, "end"]
            check_gene = True
        elif file_.loc[i,'# feature'] == 'CDS' and check_gene:
            f.loc[idx] = [name[:name.find('.')], position, position, name[:name.find('.')] + '_' + str(position), file_.loc[i, "product_accession"], file_.loc[i, "strand"]]
            idx += 1
        else:
            check_gene = False
    print(len(f))
    cords_.append(f)
```

```python
lens = []
ind = 0
for name in names:
    len_genome_ = pd.DataFrame(columns=['name', 'len'])
    gene_len = 0
    for i in SeqIO.parse(name + "_genomic.fna", "fasta"):
      gene_len += len(i.seq)

    len_genome_.loc[0] = [name[:name.find('.')], gene_len]
    lens.append(len_genome_)
```

```python
inter_dfs = {}

for name in names:
    file_ = pd.read_csv(f"intersect/{name}_inter.bed", sep="\t", names=["org_1", "start_1", "finish_1", "name_1", "zh-score", "org_2", "start_2", "finish_2", "name_2", "protein", "strand"])
    file_['start_gene'] = file_.apply(lambda x: int(x['name_2'].split('_')[-1]), axis=1)
    display(file_.head(1))
    assert(file_["org_1"].all() == file_["org_2"].all())
    inter_dfs.update({name: file_})
```

```python
def is_assotiated(row):
    for name in names:
        proteins = row[name].split(',')
        for protein in proteins:
            check = inter_dfs[name][inter_dfs[name]["protein"] == protein]
            if check.empty:
                return False
    return True
```

```python
def show_cluster(cluster, num):
  fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(40, 10))
  fig.suptitle(f'Кластер № {num}', fontsize = 25)

  for i, name in enumerate(names):
    protein = cluster[name]

    file_ = pd.read_csv(f"{name}_feature_table.txt", sep='\t')
    gene = file_[file_["product_accession"] == protein].iloc[0]

    strand = +1 if gene['strand'] == '+' else -1
    features = []

    for j, row in inter_dfs[name][inter_dfs[name]['protein'] == protein].iterrows():
      features.append(GraphicFeature(start=row['start_1'], end=row['finish_1'], strand = strand, color="#ccccff", label='Z-DNA (Score = ' + str(row['zh-score']) + ')'))
    gene_name = gene["assembly"]
    features.append(GraphicFeature(start=gene['start'], end=gene['end'], strand = strand, color="#ffcccc", label=f"gene {gene_name}"))

    l = gene['end'] - gene['start'] + 750
    f = gene['start'] - 500

    record = GraphicRecord(sequence_length=l, features=features, first_index=f)
    record.plot(ax=ax[i])
```

```python
for i, row in assotiated.iterrows():
  show_cluster(row, i + 1)
```
